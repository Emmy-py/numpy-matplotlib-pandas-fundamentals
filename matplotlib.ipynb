Sample data and structure of Matplotlib

LINE PLOT

```python
#sample data for demonstration
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)


#Line PLOT
plt.figure(figsize=(6, 4))
plt.plot(x, y, label="sin(x)", color="blue", linestyle="--", linewidth=2)
plt.title('line plot of sine wave')
plt.xlabel("x values")
plt.ylabel("sin(x)")
plt.legend()
plt.grid(True)
plt.show()


```

SCATTER PLOT

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, color="red", marker="o", alpha=0.7)
plt.title('scatter plot of scatter data')
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.show()
```

BAR PLOT

```python
categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = [10, 20, 30, 40, 50, 60]

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color="blue")
plt.title('bar chart of categories')
plt.xlabel("categories")
plt.ylabel("values")
plt.grid(True)
plt.show()
```

HISTOGRAM

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000) #normal distribution
plt.figure(figsize=(6, 4))
plt.hist(data, bins=50, color="blue", alpha=0.7, edgecolor="black")
plt.title('histogram of data')
plt.xlabel("data")
plt.ylabel("frequency")
#plt.grid(True)
plt.show()
```

Box plot

```python
# Generate random data
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(5, 1, 100)
data3 = np.random.normal(-3, 1.5, 100)

# Create boxplot
plt.figure(figsize=(6, 4))
plt.boxplot([data1, data2, data3], labels=['set1', 'set2', 'set3'])  # fixed here
plt.title('Boxplot of data')
plt.ylabel("Value")
plt.show()
```

Pie Chart

```python
slices = [40, 25, 20, 15]
labels = ['slice A', 'slice B', 'slice C', 'slice D']
explode = (0, 0.1, 0, 0)
plt.figure(figsize=(6, 4))
plt.pie(slices, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('piechart of slices')
plt.show()
```
