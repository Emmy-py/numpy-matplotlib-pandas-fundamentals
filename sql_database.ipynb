```python
import sqlite3
import pandas as pd

#create an in-memory SQLite database
connection = sqlite3.connect(':memory:')

# Create a cursor object to execute SQL commands
cursor = connection.cursor()
```

```python
#creating a table

Create_table_query = '''
CREATE TABLE Students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    age INTEGER,
    statistics_marks INTEGER,
    python_marks INTEGER,
    ml_marks INTEGER,
    dl_marks INTEGER
);
'''

cursor.execute(Create_table_query)
print("Table Created")
```

```python
#inserting data into the student table

insert_data_query = '''
INSERT INTO Students (name, age, statistics_marks,python_marks, ml_marks, dl_marks)
VALUES
('Aarav Patel', 20, 85, 90, 88, 92),
('Ayo dele', 21, 70, 60, 90, 92),
('Ngene Emmanuel', 30, 70, 80, 90, 92);

'''

cursor.execute(insert_data_query)
connection.commit()

print('data inserted into students table')
```

```python
#Query the students table and fetch data
query = 'select * from Students'
df = pd.read_sql_query(query, connection)
print("querying the students table")
df

```

```python
#update a record
update_query = "UPDATE Students SET python_marks = 95 WHERE name = 'Ngene Emmanuel'"
cursor.execute(update_query)
connection.commit()

print('updated data from students table')
```

```python
#verify the update
updated_df = pd.read_sql_query(query, connection)
updated_df
```

```python
#close the connections when done

connection.close()
cursor.close()
```

```python
query = "SELECT * FROM Students"
df = pd.read_sql(query, con=engine)
df
```

**2. filter record based on conditions**

```python
query = " SELECT * FROM Students WHERE python_marks > 85"
df = pd.read_sql(query, con=engine)
df
```

**3. Select specific columns**

```python
query = "SELECT name, statistics_marks FROM Students "
df = pd.read_sql(query, con=engine)
df
```

**4.ORDER RECORDS BY A SPECIFIC COLUMN (e.g, order by ml_marks in descending order)**

```python
query = "SELECT * FROM Students ORDER BY ml_marks DESC"
df = pd.read_sql(query, con=engine)
df
```

**5. Aggregate functions(e.g average marks in statistics**

```python
query = "SELECT AVG(statistics_marks) AS avg_statistics_marks FROM Students"
df = pd.read_sql(query, con=engine)
df
```

**6. Count the no. of students**

```python
query = "SELECT count(*) AS total_students FROM Students"
df = pd.read_sql(query, con=engine)
df
```

**7.Group by (e.g, average marks grouped by age)**

```python
query = "select age, AVG(statistics_marks) as avg_statistics_marks FROM Students GROUP BY age"
df = pd.read_sql(query, con=engine)
df
```

**8.closing SQLAlchemy connections, ost times it is closed automatically except when used by multiple users**
