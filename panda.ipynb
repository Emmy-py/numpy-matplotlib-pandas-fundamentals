```python
import random

import pandas as pd

# Create a sample dataset
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [25, 30, 35, 28, 22],
    "Department": ["Engineering", "HR", "Engineering", "Marketing", "HR"],
    "Salary": [70000, 50000, 80000, 60000, 45000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display dataset
print("Full Dataset:")
print(df)

# Task 1: Show employees older than 25
print("\nEmployees older than 25:")
print(df[df["Age"] > 25])

# Task 2: Calculate the average salary
print("\nAverage Salary:")
print(df["Salary"].mean())

# Task 3: Group employees by department and calculate average salary
print("\nAverage Salary by Department:")
print(df.groupby("Department")["Salary"].mean())

# Task 4: Sort employees by salary (highest first)
print("\nEmployees sorted by salary:")
print(df.sort_values(by="Salary", ascending=False))
```

```python
import pandas as pd
import random

import warnings
warnings.filterwarnings("ignore")

random.seed(101)

names = ['ngene','sophia', 'liam', 'jennifer']
genders = ['male', 'female', 'male', 'female']
ages = [random.randint(21,25) for _ in range(4)]
marks = [random.randint(1,10) for _ in range(4)]
subjects = random.choices(['python', 'statistics', 'generative AI'], k=4)

df = pd.DataFrame({
    'Name': names ,   # repeat names to match length
    'Gender': genders ,
    'Ages': ages,
    'Marks': marks,
    'Subjects': subjects
})

df
```
